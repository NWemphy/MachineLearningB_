```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/Dataset_UTS_Fix.xlsx")
df
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
score_df = ss.fit_transform(df)
```

```python
score_df
```

```python
df_zscore=pd.DataFrame(score_df,columns=['SL','SW', 'PL', 'PW'])
df_zscore
```

```python
data = pd.concat([df,df_zscore],axis=1)
data
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
dataS = df_zscore.values
print(dataS)

max_d = 1
dataS = df_zscore.values
plt.figure(figsize=(15,10))
plt.xlabel('sample index')
plt.ylabel('distance')
plt.title('Single Linkage')

linkage_data = linkage(dataS, method = 'single', metric = 'euclidean')
dendrogram(linkage_data)
plt.show()
```

```python
dataSingle = pd.DataFrame(df_zscore)
linkage_data = linkage(dataSingle, method = 'single', metric ='euclidean')
dendrogram(linkage_data)
plt.axhline(y=max_d, c='k', ls='--')
plt.show()

n_clusters = 3

# Membuat model Agglomerative Clustering
hierarchical_cluster = AgglomerativeClustering(n_clusters=n_clusters, metric='euclidean', linkage='single')
labels = hierarchical_cluster.fit_predict(dataSingle)

print("Hasil Clustering:", labels)
```

```python
labels = pd.DataFrame(labels,columns=['Single linkage'])
labels
```

```python
from sklearn.cluster import DBSCAN, OPTICS
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=2)
dbscan_labels = dbscan.fit_predict(dataS)
```

```python
optics = OPTICS(min_samples=2, xi=0.05, min_cluster_size=0.1)
optics_labels = optics.fit_predict(dataS)
```

```python
df_DBSCAN = pd.DataFrame(dbscan_labels,columns=['DBSCAN'])
df_DBSCAN
```

```python
df_OPTICS = pd.DataFrame(optics_labels,columns=['OPTICS'])
df_OPTICS
```

```python
data = pd.concat([df,df_zscore,labels,df_DBSCAN,df_OPTICS],axis=1)
data
```
