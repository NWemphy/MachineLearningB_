```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_excel('gaji_235314002.xlsx')
df.info()
```

```python
df=df.copy()
df
```

```python
df['USIA'] = df['USIA'].fillna(df['USIA'].mean())
df.head()
```

```python
df['GAJI']=df['GAJI'].fillna(df['GAJI'].mean())
df.tail()
```

```python
df.info()
```

```python
df.drop("NEGARA", axis=1,inplace=True)
df.drop("BELI", axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
score_df = ss.fit_transform(df)
```

```python
score_df
```

```python
df_zscore=pd.DataFrame(score_df,columns=['ZSCORE USIA','ZSCORE GAJI'])
df_zscore
```

```python
from sklearn.preprocessing import MinMaxScaler

sclaer = MinMaxScaler()
x_minmax = sclaer.fit_transform(df)
```

```python
x_minmax
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(1,10))
min_max10 = scaler.fit_transform(df)
min_max10
```

```python
df_minmax10 = pd.DataFrame(min_max10,columns=['MinMax 10 USIA','MinMax 10 GAJI'])
df_minmax10
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
min_max0 = scaler.fit_transform(df)
min_max0
```

```python
df_minmax0 = pd.DataFrame(min_max0,columns=['MinMax 0 USIA','MinMax 0 GAJI'])
df_minmax0
```

```python
data = pd.concat([df,df_zscore,df_minmax0,df_minmax10],axis=1)
data
```

```python
data.to_excel('hasil_gaji_235314002.xlsx')
```
