```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/heart_failure_clinical_records_dataset.csv")
df
```

```python
df = df.drop(columns=['diabetes'])
df
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
data = df.values
print(data)

max_d = 1
data = df.values
plt.figure(figsize=(15,10))
plt.xlabel('sample index')
plt.ylabel('distance')

linkage_data = linkage(data, method = 'single', metric = 'euclidean')
dendrogram(linkage_data)
plt.show()
plt.title('Single')

linkage_data2 = linkage(data, method = 'complete', metric ='euclidean')
dendrogram(linkage_data2)
plt.show()
plt.title('Complete')

linkage_data3 = linkage(data, method ='average', metric = 'euclidean')
dendrogram(linkage_data3)
plt.show()
plt.title('Average')

linkage_data4 = linkage(data, method = 'ward', metric = 'euclidean')
dendrogram(linkage_data4)
plt.show()
plt.title('ward')
```

```python
data = pd.DataFrame(df)
linkage_data = linkage(data, method = 'single', metric ='euclidean')
dendrogram(linkage_data)
plt.axhline(y=max_d, c='k', ls='--')
plt.show()

n_clusters = 3

# Membuat model Agglomerative Clustering
hierarchical_cluster = AgglomerativeClustering(n_clusters=n_clusters, metric='euclidean', linkage='single')
labels = hierarchical_cluster.fit_predict(data)

print("Hasil Clustering:", labels)
```
