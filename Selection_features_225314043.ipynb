```python
# Feature Selection with Univariate Statistical Tests
from pandas import read_csv
import pandas as pd
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import chi2
```

```python
# Load data
# Data berasal dari 'pima-indians-diabetes.csv' (Kaggle)
# Terdiri dari 9 atribut, 8 atribut independen dan 1 atribut/class tujuan.
# Terdiri dari 768 data observasi yang memiliki 268 positive diabetes, 500 negative diabetes
# data =read_csv('pima-indians-diabetes.csv',names=['Pregnancies','Glucose',
#    'BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Class'])
data=read_csv('/content/heart_failure_clinical_records_dataset.csv')
data.head()
```

```python
data.describe()
```

```python
array = data.values
X = array[:,0:8]
Y = array[:,8]
```

```python
X
```

```python
Y
```

```python
## Dari 8 atribut dipilih 4 atribut yang memiliki score tertinggi
## Feature extraction
test = SelectKBest(score_func=f_classif,k=3)
fit = test.fit(X, Y)
```

```python
# summarize scores
set_printoptions(precision=3)
print(fit.scores_)
features = fit.transform(X)
# summarize selected features
print(features[0:5,0:4])
```

```python
## Dari 8 atribut dipilih 4 atribut yang memiliki score tertinggi
## Feature selection
score=pd.Series(fit.scores_)
score.index=data.columns[0:8]
score.sort_values(ascending=False)
```

```python
## Ukuran data features setelah proses feature selection
features.shape
```

```python
print(fit.get_feature_names_out(input_features=data.columns[0:8]))
```

```python
# Mengubah array features yang telah diperoleh ke dalam dataframe
#df=pd.DataFrame(features,columns=['Pregnancies','Glucose','BMI','Age'])
df=pd.DataFrame(features,columns=fit.get_feature_names_out(input_features=data.columns[0:8]))
```

```python
df.head()
```

## AKURASI KNN

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectKBest, f_classif, chi2, RFE
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVR

# Load dataset
data = pd.read_csv("/content/heart_failure_clinical_records_dataset.csv")

# Pisahkan fitur (X) dan target (y)
X = data.drop(columns=["diabetes"])
y = data["diabetes"]

# Standarisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fungsi evaluasi KNN
def evaluate_knn_with_feature_selection(X, y, selector):
    X_selected = selector.fit_transform(X, y)
    X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)
    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    return accuracy_score(y_test, y_pred)

# Evaluasi berbagai metode seleksi fitur
results = {}

# SelectKBest dengan f_classif
selector_f_classif = SelectKBest(score_func=f_classif, k=3)
results["SelectKBest (f_classif)"] = evaluate_knn_with_feature_selection(X_scaled, y, selector_f_classif)

# SelectKBest dengan f_classif
selector_f_classif = SelectKBest(score_func=f_classif, k=5)
results["SelectKBest (f_classif)"] = evaluate_knn_with_feature_selection(X_scaled, y, selector_f_classif)

# SelectKBest dengan chi2 (tanpa standarisasi)
selector_chi2 = SelectKBest(score_func=chi2, k=3)
results["SelectKBest (chi2)"] = evaluate_knn_with_feature_selection(X, y, selector_chi2)

# SelectKBest dengan chi2 (tanpa standarisasi)
selector_chi2 = SelectKBest(score_func=chi2, k=5)
results["SelectKBest (chi2)"] = evaluate_knn_with_feature_selection(X, y, selector_chi2)

# RFE dengan Logistic Regression sebagai estimator
model = LogisticRegression(max_iter=1000)
selector_rfe_3 = RFE(model, n_features_to_select=3, step=1)
results["RFE (3 features)"] = evaluate_knn_with_feature_selection(X_scaled, y, selector_rfe_3)

model = LogisticRegression(max_iter=1000)
selector_rfe_5 = RFE(model, n_features_to_select=5, step=1)
results["RFE (5 features)"] = evaluate_knn_with_feature_selection(X_scaled, y, selector_rfe_5)

# RFE dengan SVR sebagai estimator
model_svr = SVR(kernel="linear")
selector_rfe_svr_3 = RFE(model_svr, n_features_to_select=3, step=1)
results["RFE (3 features) - SVR"] = evaluate_knn_with_feature_selection(X_scaled, y, selector_rfe_svr_3)

selector_rfe_svr_5 = RFE(model_svr, n_features_to_select=5, step=1)
results["RFE (5 features) - SVR"] = evaluate_knn_with_feature_selection(X_scaled, y, selector_rfe_svr_5)

# Cetak hasil akurasi
sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True)
print("Ranking Akurasi:")
for rank, (method, accuracy) in enumerate(sorted_results, start=1):
    print(f"{rank}. {method}: {accuracy:.4f}")
```
