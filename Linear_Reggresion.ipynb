```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Import data
dataku = pd.read_csv('Daftar_gaji.csv')
```

```python
#3 Menampilkan isi dataset

#dataku=pd.DataFrame(dataset)
#dataku

#Visualisasi Data
plt.scatter(dataku.Tahun_bekerja,dataku.Gaji)
plt.xlabel("Tahun")
plt.ylabel("Pendapatan")
plt.title("Masa Kerja vs Pendapatan")
plt.show()

dataku
```

```python
from sklearn.model_selection import train_test_split
X = dataku.iloc[:,:-1].values
y = dataku.iloc[:,1].values
#print(X)
#print(y)

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
r_sq=model.score(X_train,y_train)
print(f"Cooefficient of determination:  {r_sq}")
```

```python
print(f"intercept:  {model.intercept_}")
```

```python
print(f"coefficients: {model.coef_}")
```

```python
# Visualisasi hasil Training Set
plt.scatter(X_train, y_train,color ='red')
plt.plot(X_train, model.predict(X_train),color='blue')
plt.title('Gaji vs Pengalaman (Training set)')
plt.xlabel('Tahun bekerja')
plt.ylabel('Gaji')
plt.show()
```

```python
# 7 Visualisasi hasil Test Set
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.title("Masa Kerja vs Pendapatan (Training set)")
plt.xlabel("Tahun bekerja")
plt.ylabel("Pendapatan")
plt.show()
```
